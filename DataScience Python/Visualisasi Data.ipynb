```python
#Import package
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#Membuat scatterplot
a=[2,5,6,7,4,3,1]
b=[9,5,6,8,7,3,2]

plt.scatter(a,b)
plt.show()
```

```python
#Histogram dengan seaborn
sns.histplot(a)
plt.show()
```

```python
#Membuat lineplot dengan Matplotlib
plt.title("Perbandingan Nilai Fisika Siswa dan Lama Belajar Siswa (jam)")
plt.xlabel('Nilai Fisika')
plt.ylabel('Lama belajar')
plt.plot(a,b)
```

```python
#Membuat scatterplot dengan costumisasi
c=[2,5,6,7,4,3,1]
d=[9,5,6,8,7,3,2]

plt.scatter(c,d, color='yellow', marker='v')
plt.show()
```

```python
data1=np.random.randn(5000)
plt.hist(data1, bins=30, color='blue')
plt.title("Histogram")
plt.xlabel("Nilai")
plt.ylabel("Frekuensi")
plt.show()
```

```python
#Membuat Barplot
Kelompok=['A','B','C','D','E','F']
Nilai=[10,15,19,13,16,17]
plt.bar(Kelompok,Nilai, color='red')
plt.title("Nilai Matematika Masing-masing Kelompok Kelas VIIA")
plt.xlabel("Kelompok")
plt.ylabel("Nilai")
```

```python
#Membuat Pie Chart #nilai dan label tidak perlu dikasih tanda petik
plt.pie(Nilai, labels=Kelompok, colors=('red','green','blue','grey','orange','pink'), autopct='%1.1f%%')
plt.title('Diagram lingkaran Nilai Matematika Kelompok Kelas VIIA')
```

```python
#Membuat Boxplot
data2=[np.random.normal(0,std,200) for std in range(1,5)]
plt.boxplot(data2, labels=('A','B','C','D'))

plt.title('Boxplot')
plt.xlabel('Kategori')
plt.ylabel('Nilai')
plt.show()
```

```python
data3=np.random.rand(6,6) #membuat heatmap
plt.imshow(data3, cmap='hot')
plt.colorbar()
plt.title('Heatmap')
plt.show()
```

```python
import pandas as pd
Data3=pd.DataFrame(data3)
sns.pairplot(Data3)
```

```python
#Membuat Boxplot untuk perbandingan
data4=({'Nama':('Andi','Budi','Cahya','Doni','Erik'),
       'Kelas':(10,11,12,11,12),
      'Warna Kesukaan':('Kuning','Biru','Merah','Kuning','Biru'),
       'Nilai Fisika':(100,93,96,99,92)})
Data4=pd.DataFrame(data4)
Data4
```

```python
#Perbandingan antara kelas dan warna kesukaan (jangan seperti ini, terbalik)
sns.boxplot(x='Kelas',y='Warna Kesukaan',data=Data4)
plt.title=('Perbandingan antara kelas dan warna kesukaan')
plt.xlabel('Kelas')
plt.ylabel('Warna Kesukaan')
plt.show()
```

```python
#Perbandingan antara kelas dan warna kesukaan dengan boxplot
sns.boxplot(x='Warna Kesukaan',y='Kelas',data=Data4)
plt.title=('Perbandingan antara kelas dan warna kesukaan')
plt.xlabel('Warna Kesukaan') #variabel kategorikal
plt.ylabel('Kelas') #variabel berupa nilai
plt.show()
```

```python
#Violinplot
sns.violinplot(x='Warna Kesukaan',y='Kelas',data=Data4)
plt.title=('Perbandingan antara kelas dan warna kesukaan')
plt.xlabel('Warna Kesukaan') #variabel kategorikal
plt.ylabel('Kelas') #variabel berupa nilai
plt.show()
```
